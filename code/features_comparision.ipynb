```python


#from helper import min_max_range

data_path = '/Users/mylesgould/Downloads/project/data/'
labels = ['rotate_next', 'rotate_prev', 'like', 'stop']#, 'other']

raw_data_types = ['accel_x', 'accel_y', 'accel_z', 'rotation_x', 'rotation_y', 'rotation_z', 'roll', ' pitch', 'yaw']



# Collect all features for all labels and raw_data_types into a DataFrame and save as CSV
all_features = []

for label in labels:
        folder_path = os.path.join(data_path, label, 'csv')
        for csv_path in os.listdir(folder_path):
            df = pd.read_csv(os.path.join(folder_path, csv_path))
            if raw_data not in df.columns:
                continue

            features_storage = {'label': label, 'file': csv_path}
            # Initialize features for each raw data type
            for raw_data in raw_data_types:
                if raw_data not in df.columns:
                    continue
                signal = df[raw_data].values

                extracted_features = {

                    f'{raw_data}_std': std(signal),
                    f'{raw_data}_energy': energy(signal),
                    f'{raw_data}_mean': np.mean(signal),
                    f'{raw_data}_max': max_val(signal),
                    f'{raw_data}_min': min_val(signal),
                    # 'min_max_range': min_max_range(signal),
                    f'{raw_data}_mam': median_above_mean(signal),
                    f'{raw_data}_itqr': interquartile_range(signal),
                    f'{raw_data}_skewness': skewness(signal)

                }
                extracted_psd = psd_features(signal)

                renamed_extracted_psd = {f'{raw_data}_{k}': v for k, v in extracted_psd.items()}
                
                features_storage.update(extracted_features)
                features_storage.update(renamed_extracted_psd)
            all_features.append(features_storage)  
                


# Convert to DataFrame and save
features_df = pd.DataFrame(all_features)
features_df.to_csv(os.path.join(os.getcwd(), 'all_features.csv'), index=False)
```

```python
import torch
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split

# Load a CSV file into a DataFrame
csv_path = 'all_features.csv'  # Change this to your actual file path

csv_path = os.path.join(os.getcwd(), csv_path)
if not os.path.exists(csv_path):
    raise FileNotFoundError(f"The file {csv_path} does not exist.")

print(f"Loading CSV file from: {csv_path}")
df = pd.read_csv(csv_path)

# Get the labels (assuming a column named 'label')
labels = df['label'].values

# Get the data for each row (excluding the label and file columns)
data_columns = [col for col in df.columns if col not in ['label', 'file']]
data = df[data_columns].values

#Encode labels as integers
label_encoder = LabelEncoder()
labels_encoded = label_encoder.fit_transform(labels)

#Standardize the data
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)

# Split the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(
    data_scaled, labels_encoded, test_size=0.3, random_state=42, stratify=labels_encoded
)

# Convert to PyTorch tensors
X = torch.tensor(X_train, dtype=torch.float32)
y = torch.tensor(y_train, dtype=torch.long)
X_test = torch.tensor(X_test, dtype=torch.float32)
y_test = torch.tensor(y_test, dtype=torch.long)

print("X shape:", X.shape)
print("y shape:", y.shape)
print("Classes:", label_encoder.classes_)

# Simple neural network using torch.nn
import torch.nn as nn
import torch.optim as optim

class SimpleNN(nn.Module):
    def __init__(self, input_dim, num_classes):
        super(SimpleNN, self).__init__()
        self.fc1 = nn.Linear(input_dim, 32)
        self.relu = nn.ReLU()
        self.fc2 = nn.Linear(32, num_classes)
    
    def forward(self, x):
        x = self.fc1(x)
        x = self.relu(x)
        x = self.fc2(x)
        return x

input_dim = X.shape[1]
num_classes = len(label_encoder.classes_)
model = SimpleNN(input_dim, num_classes)

criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.01)

# Training loop (very basic, for demonstration)
epochs = 40
for epoch in range(epochs):
    optimizer.zero_grad()
    outputs = model(X)
    loss = criterion(outputs, y)
    loss.backward()
    optimizer.step()

    if (epoch + 1) % 10 == 0:
        model.eval()
        with torch.no_grad():
            outputs_test = model(X_test)
            _, predicted = torch.max(outputs_test, 1)
            accuracy = (predicted == y_test).float().mean().item()
            print(f"Epoch [{epoch+1}/{epochs}], Loss: {loss.item():.4f}, Test Accuracy: {accuracy:.4f}")
        model.train()
    else:
        print(f"Epoch [{epoch+1}/{epochs}], Loss: {loss.item():.4f}")
```

```python
# Save the trained PyTorch model to a file
model_save_path = os.path.join(os.getcwd(), 'model.pth')
torch.save(model.state_dict(), model_save_path)
print(f"Model saved to {model_save_path}")
```
